```python
import scipy.integrate as spi
import numpy as np
import matplotlib.pyplot as plt

# N总数
N=10000
beta=0.25
gammga=0.05
I_0=1
R_0=0
S_0=N-I_0-R_0
T=150
INI=(S_0,I_0,R_0)

def fucSI(inivalue,_):
    Y=np.zeros(3)
    X=inivalue
    Y[0]=-(beta*X[0]*X[1])/N
    Y[1]=(beta*X[0]*X[1]/N-gammga*X[1])
    Y[2]=gammga*X[1]
    return Y
T_range=np.arange(0,T+1)

RES=spi.odeint(fucSI,INI,T_range)

plt.plot(RES[:,0],color="darkblue",label="Susceptible",marker=",")
plt.plot(RES[:,1],color="red",label="Infection",marker=",")
plt.plot(RES[:,2],color="green",label="Recovery",marker=",")
plt.title("SIR Model")
plt.legend()
plt.xlabel("Day")
plt.ylabel("Number")
plt.show()
```
